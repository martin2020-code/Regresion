```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
model = lambda t,*p: p[0]*t*np.exp(p[1]*t)
t = np.linspace(0,10)
y = model(t,0.1,-0.5) + np.random.randn(t.size)*2e-3


delta = 0.1
iter = 1
p0 = [0.1,-0.5]
for i in range(iter):
    p0[0] = p0[0] + 2*delta*np.sum(model(t,*p0)*t*np.exp(-p0[1]*t))
    p0[1] = p0[1] - 2*delta*np.sum(model(t,*p0)*t*p0[0]*p0[1])


pfit, pcov = curve_fit(model, t,y,p0=[0.1,-0.5])
plt.plot(t,y,'.')
plt.plot(t, model(t,*pfit),'--')
p0
```

```python
poly = lambda t: 2*t + 3*t**2 - 0.5*t**3
der = lambda t: 2 + 6*t - 1.5*t**2
plt.plot(t,poly(t))
plt.plot(t,der(t))
```
